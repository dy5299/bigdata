```python
import apyori as ap
help(ap)
```

```python
tr = [['우유', '계란,' '커피'],
      ['빵', '우유', '버터', '휴지', '라면'],
      ['맥주', '땅콩'],
      ['계란', '빵', '우유', '생수'],
      ['햇반', '커피', '간장', '계란']]
```

```python
rules = ap.apriori(tr, min_support=0.3, min_confidence=0.7)
```

```python
for r in rules:
    print(r)
```
