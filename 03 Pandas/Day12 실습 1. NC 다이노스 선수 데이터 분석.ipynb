# NC 다이노스 선수 데이터 분석
## 1. 데이터 적재 및 통합

```python
data = pd.read_excel('data/NC DInos.xlsx', sheet_name = None)
NC13, NC14, NC15 = data.values()
```

```python
# 년도 컬럼 추가
NC13['년도'] = 2013
NC14['년도'] = 2014
NC15['년도'] = 2015
```

```python
# pandas version >= 0.24인 경우 Unnamed: 0 라는 컬럼이 있음.
# 있는 경우, 아래 구문으로 삭제
NC13.drop('Unnamed: 0', axis=1, inplace = True)
NC14.drop('Unnamed: 0', axis=1, inplace = True)
NC15.drop('Unnamed: 0', axis=1, inplace = True)
```

```python
pd.__version__
```

```python
NC13.head()
```

```python
# 하나의 데이터프레임으로 합치기
NC = pd.concat([NC13, NC14 ,NC15], axis=0, ignore_index = True)
NC.head()
```

## (실습 #2) 2013~2015년 3년 동안 선수들의 평균 안타수, 총 안타수, 선수로 뛴 시즌수 구하기
선수로 뛴 시즌수 = 해당 시즌의 선수 목록에 없다면 그 선수는 해당 시즌에 선수에 포함되지 않은 것임. 

```python
tb1 = NC.pivot_table(index='선수명', values='안타', aggfunc='mean')
# .rename(columns={'안타':'평균 안타수'})
tb1.columns = ['평균 안타수']
tb2 = NC.pivot_table(index='선수명', values='안타', aggfunc='sum')
tb2.columns = ['총 안타수']
tb3 = NC.pivot_table(index='선수명', values='안타', aggfunc='count')
tb3.columns = ['시즌수']
Q2 = pd.concat([tb1, tb2, tb3])
# .set_columns(['평균 안타수', '총 안타수', '시즌수'])
Q2
```

```python
# {'평균안타수':np.mean, '총안타수':np.sum, '시즌수':len}
```

### Ans

```python
Q2 = NC.pivot_table(index = '선수명', values = '안타', aggfunc = ['mean', 'sum', 'count'])
# count() 함수로 row의 개수를 구할 떄에는 NaN이 없는 임의의 컬럼을 이용하면 됩니다.
Q2.head()
```

```python
Q2.columns = ['평균안타', '총안타', '참여시즌수']
Q2
```
